```python
import pandas as pd
```

```python
df = pd.read_csv("../data/raw/HR-Employee-Attrition.csv")
```

```python
#Drop Constant/useless columns
df.drop(columns=['EmployeeCount', 'StandardHours', 'Over18', 'EmployeeNumber'], inplace = True, errors = 'ignore')
```

```python
#Encode target
df['Attrition_Flag'] = df['Attrition'].map({'Yes':1, 'No':0})
```

```python
# One-hot encode categoricals for modeling later
cat_cols = df.select_dtypes(include=['object', 'string']).columns.tolist()
cat_cols.remove('Attrition')
df_encoded = pd.get_dummies(df, columns=cat_cols, drop_first=True)
```

```python
df.to_csv("../data/processed/hr_cleaned.csv", index = False)
df_encoded.to_csv("../data/processed/hr_encoded.csv", index=False)
print("Saved Cleaned + encoded version")
```
