```python
plt.style.use("ggplot")
```

```python
# Attrition by department
plt.figure(figsize=(8,5))
sns.countplot(data=df, x='Department', hue='Attrition')
plt.title('Attrition by department')
plt.savefig("../images/attrition_VS_department.png",
            dpi=300,
            bbox_inches="tight")
plt.show()
```

```python
#Overtime Vs Attrition - Usually the strongest signal in this dataset
sns.countplot(data=df, x='OverTime', hue='Attrition')
plt.title('Attrition vs OverTime')
plt.savefig("../images/Overtime_VS_attrition.png",
            dpi=300,
            bbox_inches="tight")
plt.show()
```

```python
# Income distribution by attrition
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome')
plt.title('Monthly Income Vs Attrition')
plt.savefig("../images/IncomeDist_VS_attrition.png",
            dpi=300,
            bbox_inches="tight")
plt.show()
```

```python
# Correlation heatmap (numeric only)
plt.figure(figsize=(14,10))
sns.heatmap(df.corr(numeric_only=True), cmap='coolwarm', annot = False)
plt.title('Correlation Heatmap')
plt.savefig("../images/Correlation_Heatmap.png",
            dpi=300,
            bbox_inches="tight")
plt.show()
```

```python
# Age distribution
sns.histplot(data = df, x='Age', hue='Attrition', kde=True, multiple='stack')
plt.savefig("../images/Age_Distribution.png",
            dpi=300,
            bbox_inches="tight")
plt.show()
```

```python
# Job Role Vs Attrition
sns.countplot(data=df, x='JobRole', hue='Attrition')
plt.title('JobRole vs Attrition')
plt.xlabel('Job Role')
plt.ylabel('Number of Employees')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
#Job Satisfaction vs Attrition
sns.countplot(data=df, x='JobSatisfaction', hue='Attrition')
plt.title("Job Satisfaction Vs Attrition")
plt.xlabel('Job Satisfaction')
plt.ylabel('Number of Employees')
plt.show()
```

```python
#Work Life Balance vs Attrition
sns.countplot(data = df, x = 'WorkLifeBalance', hue = 'Attrition')
plt.title('Work Life Balance VS Attrition')
plt.xlabel('Work Life Balance')
plt.ylabel('No. of Employees')
plt.show()
```

```python
#Environment Satisfaction vs Attrition
sns.countplot(data = df, x='EnvironmentSatisfaction', hue='Attrition')
plt.title('Environment Satisfaction Vs Attrition')
plt.xlabel('Environment Satisfaction')
plt.ylabel('Number of Employees')
plt.show()
```

```python
#Monthly Income by Department
sns.boxplot(data = df, x= 'Department', y='MonthlyIncome')
plt.title('Monthly Income Vs Department')
plt.xlabel('Department')
plt.ylabel('Monthly Income')
plt.show()
```
